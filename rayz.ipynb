```python
import os
import shutil
from tqdm import tqdm
# Paths
base_dir = 'nhih'  #
output_dir = 'normalized_images'  
os.makedirs(output_dir, exist_ok=True)
# Collect all images
folders = [f"images_{str(i).zfill(3)}" for i in range(1, 13)] 

for folder in tqdm(folders):
    folder_path = os.path.join(base_dir, folder, 'images') 

    for img_file in os.listdir(folder_path):
        src_path = os.path.join(folder_path, img_file)

        if os.path.isfile(src_path):
            dest_path = os.path.join(output_dir, img_file)
            shutil.copy(src_path, dest_path)
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
input_dir = 'normalized_images'  
output_dir = '/content/normalized_images_final' 
os.makedirs(output_dir, exist_ok=True)
def normalize_image(img_path):
    try:
        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        image = cv2.resize(image, (224, 224))  # Resize
        image = (image - np.min(image)) / (np.max(image) - np.min(image))  # Min-Max Normalize
        output_path = os.path.join(output_dir, os.path.basename(img_path))
        cv2.imwrite(output_path, (image * 255).astype(np.uint8))  # Save normalized image
    except Exception as e:
        print(f"Error processing {img_path}: {e}")

img_files = [os.path.join(input_dir, f) for f in os.listdir(input_dir)]
with ThreadPoolExecutor(max_workers=8) as executor:  # Multi-threading
    list(tqdm(executor.map(normalize_image, img_files), total=len(img_files)))

print("Normalization complete!")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import os
import numpy as np
import cv2

classes = [
    "No Finding", "Atelectasis", "Cardiomegaly", "Effusion", "Infiltration",
    "Mass", "Nodule", "Pneumonia", "Pneumothorax", "Consolidation",
    "Edema", "Emphysema", "Fibrosis", "Pleural_Thickening", "Hernia"
]

def encode_labels(label_str):
    labels = np.zeros(len(classes))
    for label in label_str.split('|'):
        if label in classes:
            labels[classes.index(label)] = 1
    return labels

default_image = np.zeros((3, 224, 224), dtype=np.float32)

class ChestXRayDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform
        self.samples = list(zip(self.image_paths, self.labels))

    def __len__(self):
        return len(self.samples)  

    def __getitem__(self, idx):
        image_path, label_str = self.samples[idx]

        image = cv2.imread(image_path)
        if image is None:
            print(f"Error loading image: {image_path}")
            image = default_image  
        else:
            image = cv2.resize(image, (224, 224))
            image = np.transpose(image, (2, 0, 1)) / 255.0  # Normalize

        label = encode_labels(label_str)
        # Convert to tensor
        image = torch.tensor(image, dtype=torch.float32)
        label = torch.tensor(label, dtype=torch.float32)
        return image, label


csv_file = '/content/Data_Entry_2017.csv'
df = pd.read_csv(csv_file)
df = df[['Image Index', 'Finding Labels']]
image_paths = [os.path.join("/content/nif", img) for img in df['Image Index']]
labels = df['Finding Labels'].tolist()
dataset = ChestXRayDataset(image_paths=image_paths, labels=labels)
train_loader = DataLoader(dataset, batch_size=64, shuffle=True, num_workers=4, pin_memory=True)

print(" Dataset Loaded Successfully!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import os
import torch.nn.utils.prune as prune
import torch_tensorrt as torchtrt
from torch.ao.quantization import quantize_dynamic
from torch.utils.data import DataLoader, Subset
from torchvision import models
from torchvision.models import ConvNeXt_Base_Weights, EfficientNet_B3_Weights, ResNet18_Weights, DenseNet121_Weights, ViT_B_16_Weights, Swin_V2_S_Weights
from sklearn.model_selection import train_test_split
from PIL import Image
import optuna

os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# Device 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True
torch.cuda.empty_cache()
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True

# Mixed Precision Training
scaler = torch.amp.GradScaler() if torch.cuda.is_available() else None

# Configuration
num_classes = 15
BATCH_SIZE = 32
NUM_WORKERS = 4  
NUM_EPOCHS = 5
EARLY_STOPPING_PATIENCE = 3

# DataLoader
def get_dataloader(dataset, batch_size, shuffle=True):
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, num_workers=NUM_WORKERS, pin_memory=True, persistent_workers=True)

def verify_image(path):
    """Ensures the image is valid before loading"""
    try:
        with Image.open(path) as img:
            img.verify()
        return True
    except:
        return False


dataset.samples = [(path, label) for path, label in dataset.samples if verify_image(path)]
train_idx, test_idx = train_test_split(list(range(len(dataset))), test_size=0.2, random_state=42)
train_idx, val_idx = train_test_split(train_idx, test_size=0.25, random_state=42)
train_dataset = Subset(dataset, train_idx)
val_dataset = Subset(dataset, val_idx)
test_dataset = Subset(dataset, test_idx)
train_loader = get_dataloader(train_dataset, batch_size=BATCH_SIZE)
val_loader = get_dataloader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)


models_to_test = {
    "ResNet18": models.resnet18(weights=ResNet18_Weights.IMAGENET1K_V1),
    "DenseNet121": models.densenet121(weights=DenseNet121_Weights.IMAGENET1K_V1),
    "EfficientNet_B3": models.efficientnet_b3(weights=EfficientNet_B3_Weights.IMAGENET1K_V1),
    "ConvNeXt_Base": models.convnext_base(weights=ConvNeXt_Base_Weights.IMAGENET1K_V1),
    "ViT_Base": models.vit_b_16(weights=ViT_B_16_Weights.IMAGENET1K_V1),
    "Swin_V2_S": models.swin_v2_s(weights=Swin_V2_S_Weights.IMAGENET1K_V1),
}

#  Multi-Class Classification
def fine_tune_model(model):
    for param in model.parameters():
        param.requires_grad = False 

    if hasattr(model, 'fc'):
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif hasattr(model, 'classifier'):
        if isinstance(model.classifier, nn.Linear):
            model.classifier = nn.Linear(model.classifier.in_features, num_classes)
        elif isinstance(model.classifier, nn.Sequential):
            model.classifier[-1] = nn.Linear(model.classifier[-1].in_features, num_classes)
    elif hasattr(model, 'heads'):
        model.heads.head = nn.Linear(model.heads.head.in_features, num_classes)
    elif hasattr(model, 'head'):
        model.head = nn.Linear(model.head.in_features, num_classes)

    for param in model.parameters():
        param.requires_grad = True 
    return model


def evaluate(model, data_loader, criterion):
    model.eval()
    total_loss = 0.0
    total_correct, total_samples = 0, 0

    with torch.no_grad():
        for images, labels in data_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            predictions = (outputs > 0).int()
            total_correct += (predictions == labels).sum().item()
            total_samples += labels.numel()

    accuracy = 100 * total_correct / total_samples
    return total_loss / len(data_loader), accuracy


def train_and_evaluate(model, model_name, criterion, optimizer, train_loader, val_loader):
    model.to(device)
    best_loss = float('inf')
    patience = 0

    for epoch in range(NUM_EPOCHS):
        model.train()
        running_loss = 0.0
        optimizer.zero_grad()

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            with torch.amp.autocast(device_type="cuda", dtype=torch.float16):
                outputs = model(images)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
            running_loss += loss.item()

        avg_loss = running_loss / len(train_loader)
        val_loss, accuracy = evaluate(model, val_loader, criterion)

        print(f"Model: {model_name}, Epoch {epoch+1}, Loss: {avg_loss:.4f}, Val Loss: {val_loss:.4f}, Accuracy: {accuracy:.2f}%")

        if val_loss < best_loss:
            best_loss = val_loss
            patience = 0
        else:
            patience += 1
            if patience >= EARLY_STOPPING_PATIENCE:
                print(f"Early stopping for {model_name} at Epoch {epoch+1}")
                break

    return accuracy, model

# Optuna Hyperparameter Search
def objective(trial):
    model_name = trial.suggest_categorical("model_name", list(models_to_test.keys()))
    lr = trial.suggest_float("lr", 1e-5, 1e-2, log=True)

    torch.cuda.empty_cache()

    try:
        model = fine_tune_model(models_to_test[model_name])
        optimizer = optim.Adam(model.parameters(), lr=lr)
        criterion = nn.BCEWithLogitsLoss()
        accuracy, _ = train_and_evaluate(model, model_name, criterion, optimizer, train_loader, val_loader)
        return accuracy
    except torch.cuda.OutOfMemoryError:
        return 0

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=5)
best_model_name = study.best_params["model_name"]

# Convert to TensorRT
best_model = fine_tune_model(models_to_test[best_model_name]).to(device)
scripted_model = torch.jit.script(best_model)
scripted_model.save(f"{best_model_name}_optimized_trt.ts")

print(f" Best Model {best_model_name} Saved & Optimized with TensorRT!")
```

```python
import torch
import torch.nn.utils.prune as prune
from torch.ao.quantization import quantize_dynamic
model = torch.jit.load(f"{best_model_name}_optimized_trt.ts").to(device)
for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d) or isinstance(module, torch.nn.Linear):
        prune.l1_unstructured(module, name="weight", amount=0.3)

quantized_model = quantize_dynamic(model, {torch.nn.Linear}, dtype=torch.qint8)
torch.jit.save(quantized_model, f"{best_model_name}_pruned_quantized.ts")
print(f" {best_model_name} Pruned & Quantized Model Saved!")
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image
model = torch.jit.load(f"{best_model_name}_pruned_quantized.ts").to(device)
model.eval()

image_path = "/content/nif/00000003_000.png"  
image = Image.open(image_path).convert("RGB")  

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
input_tensor = transform(image).unsqueeze(0).to(device)

if input_tensor.shape[1] == 1:
    input_tensor = input_tensor.repeat(1, 3, 1, 1)

with torch.no_grad():
    output = model(input_tensor)

prediction = torch.argmax(output, dim=1)
print(f" Model Prediction (Fixed): {prediction.item()}")
```
